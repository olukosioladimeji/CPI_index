```python
import tensorflow as tf
import numpy as np
```

```python
year = np.array([2012,2013,2014,2015,2016,2017,2018,2019,2020,2021],dtype=float)
cpi_rank = np.array([27,25,27,26,28,27,27,26,25,154],dtype=float)
for i,c in enumerate(year):
  print("{} year = {} cpi_rank".format(c,cpi_rank[i]))
```

```python
l0 = tf.keras.layers.Dense(units=1,input_shape=[1])
```

```python
model=tf.keras.Sequential([l0])
```

```python
model.compile(loss='mean_squared_error',optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history = model.fit(year,cpi_rank,epochs=500,verbose=False)
print("we are done training our model")
```

```python
print(model.predict([2022]))
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Epoch Number')
plt.ylabel('Loss Magnitude')
plt.plot(history.history['loss'])
```
